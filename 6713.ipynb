```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

bus_data = pd.read_csv('BUS_STATION_BOARDING_MONTH_202211.csv',engine='python',encoding='CP949')
```

```python
bus_data.info()
```

```python
bus_data.head()
```

```python
#6713번 노선 데이터를 추출
#6713번 노선의 정류장별 승하차량 추출
bus_6713 = bus_data[bus_data['route'] == "6713"]
bus_6713 = bus_6713.groupby(["route", "stop_id", "stop_name"])\
        .agg(total_board = ('board', 'mean'), total_off = ('deboard', 'mean')).reset_index()

bus_6713['total_board'] = bus_6713['total_board'].astype(int)
bus_6713['total_off'] = bus_6713['total_off'].astype(int)
bus_6713['total_usage'] = bus_6713['total_board'] + bus_6713['total_off']

bus_6713.info()
```

```python
bus_6713.head()
```

```python
bus_6713.to_csv('Bus_6713_boarding_data_202211.csv', encoding='CP949')
```

```python
bus_6713.sort_values(by=["total_usage"], ascending=False).head(30)
```
