```python
import pandas as pd
import numpy as np

df = pd.read_csv('BUS_STATION_BOARDING_MONTH_202211.csv',engine='python',encoding='CP949')
```

```python
route_info = pd.read_excel('서울시 버스노선 정보(20220822).xlsx')
route_info.head()
```

```python
df2 = df.groupby(['date', 'route'])\
.agg(total_board = ('board', 'sum')).reset_index()
```

```python
bus_data = pd.merge(left = df2 , right = route_info, how = "left", on = "route")
bus_data.head()
```

```python
df3 = bus_data.groupby(['route', "Route_type","n_bus"])\
.agg(total_board = ('total_board', 'mean')).reset_index()

df3['total_board'] = df3['total_board'].astype(int)
df3['n_bus'] = df3['n_bus'].astype(int)

df3 = df3.sort_values(by=['total_board'], ascending=False)

df3['board_div_n_bus'] = df3['total_board'] / df3['n_bus']
df3['board_div_n_bus'] = df3['board_div_n_bus'].astype(int)

df3_n_bus = df3.sort_values(by=['board_div_n_bus'], ascending=False)

df3.info()
```

```python
df3.head()
```

```python
df3.to_csv('Bus_boarding_data_nov.csv', encoding='CP949')
df3_n_bus.to_csv('Bus_boarding_data_by_nbus_nov.csv', encoding='CP949')
```
